```python
# вдруг пригодится
data=data.replace({-1: 0})
data

col_num=data.shape[1]
row_num=data.shape[0]
#print(col_num,row_num)
```

```python
Здесь главная функция, считающая метрику сходства
```

```python
def sim_func(user_num):
    Mat = np.zeros((row_num,row_num)) #нужно внимательно понять, что в конечном итоге нам нужно будет 38 столбцов 
    for userU in range(row_num):
        for userV in range(row_num):
            if(userU != userV):
                A = data.iloc[lambda df: [userU,userV]]
                ASQRT = (data.iloc[lambda df: [userU,userV]]) ** 2
                MatrSum = ASQRT.sum(axis = 1) #матрица общей суммы квадратов
                pr_znam = m.sqrt(MatrSum[userU] * MatrSum[userV])
                sum_p = 0;
                chisl=0;
                for film in range(col_num):
    #                 chisl+=data[film][userU] * data[film][userV]
                    user_film = A[A.columns[film]]
                    sum_p += user_film[userU] * user_film[userV] #даже если 0, то будет 0 и он не повлияет на сумму
                Mat[userU,userV] = round(sum_p/ pr_znam,3)
    return(Mat[user_num])
print("Метрика сходства для нашего пользователя:\n",sim_func(20))
```

```python
#далее работаем с этими  пользователями
ru=round(ru,3)
sm = np.zeros(col_num) 
sum_m = 0 #сумма метрик сходства пользователей
for us in index:
    dU = data.iloc[[us]] # берем данные по фильмам у требуемого пользователя
    #Вызовем нашу функцию для поиска среднего по пользователю v
    srv=rv(us)
    sum_m = round(sum_m + sim_func(20)[us], 3)
    for film in range(col_num):
        sm[film] =  sm[film] + sim_func(20)[us] * (dU[dU.columns[film]] - ru)
for i in range(col_num):
    sm[i] = round(srv + sm[i] / sum_m,3)
print("Новые оценки для 21 User:", sm)
```
